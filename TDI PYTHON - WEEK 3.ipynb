```python
#Calculating Average Fare
#Write a script to calculate the average fare paid by passengers using the Titanic dataset.
```

```python
# Calculating the average fare from Titanic dataset

import csv

with open("TDI python week3 dataset(Titanic Analysis).csv", mode="r") as file:
    csv_reader = csv.DictReader(file)  # Use DictReader so we can access by column names
    
    total_fare = 0.0
    passenger_count = 0
    
    for row in csv_reader:
        if row["fare"]:  # Make sure the value isn’t empty
            total_fare += float(row["fare"])
            passenger_count += 1
    
    # Calculate average fare
    average_fare = total_fare / passenger_count if passenger_count > 0 else 0

print(f"Average fare paid by passengers: {average_fare:.2f}")
```

```python
#Survivors by Class
#Write a Python script to calculate the number of survivors per passenger class (1st, 2nd, and 3rd class).
```

```python
import csv

# Open the Titanic dataset
with open("TDI python week3 dataset(Titanic Analysis).csv", mode="r") as file:
    csv_reader = csv.DictReader(file)  # Use DictReader to access columns by name
    
    # Dictionary to store survivors per class
    survivors_by_class = {"1": 0, "2": 0, "3": 0}
    
    # Loop through each row
    for row in csv_reader:
        if row["survived"] == "1":  # Check if passenger survived
            pclass = row["pclass"]  # Get passenger class
            survivors_by_class[pclass] += 1  # Increment survivor count for that class
    
    # Print results
    for pclass, count in survivors_by_class.items():
        print(f"Survivors in Class {pclass}: {count}")
```

```python
#Titanic Data with NumPy
#Convert the Fare column of the Titanic dataset to a NumPy array and
#calculate:
#• The minimum fare.
#• The maximum fare.
#• The mean fare.
```

```python
import csv
import numpy as np

# Read Titanic dataset and extract 'fare' column
fares = []
with open("TDI python week3 dataset(Titanic Analysis).csv", "r") as file:
    csv_reader = csv.DictReader(file)
    for row in csv_reader:
        if row["fare"]:  # check if not empty
            fares.append(float(row["fare"]))

# Convert to NumPy array
fare_array = np.array(fares)

# Calculate statistics
min_fare = np.min(fare_array)
max_fare = np.max(fare_array)
mean_fare = np.mean(fare_array)

print("Minimum Fare:", round(min_fare, 2))
print("Maximum Fare:", round(max_fare, 2))
print("Mean Fare:", round(mean_fare, 2))
```
